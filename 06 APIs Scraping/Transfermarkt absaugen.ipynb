### 1. Bibliotheken laden

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### 2. Abfrage starten

```python
final_data = []
```

```python
for saison in list(range(2008, 2019)):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    url = 'https://www.transfermarkt.ch/1-bundesliga/startseite/wettbewerb/L1/saison_id/'+str(saison)
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    # Vereine filtern
    vereinsliste = []
    vereine = soup.find_all('td', {'class':'hauptlink no-border-links show-for-small show-for-pad'})
    for verein in vereine:
        vereinsliste.append(verein.text)

    # Marktwerte filten
    marktwerte = soup.find_all('td', {'class':'rechts show-for-small show-for-pad nowrap'})
    marktwertliste = []
    for wert in marktwerte[2:][::2]: # Zwei abschneiden, dann nur jedes zweite Element
        marktwertliste.append(float(wert.text.replace(' Mio. €', '').replace(',', '.')))

    # Zu einer gemeinsamen Liste zusammenfügen
    for verein, wert in zip(vereinsliste, marktwertliste):
        paar = {"Verein": verein,
               "Wert": wert,
               "Saison": saison}
        final_data.append(paar)        
```

### 3. Mit Pandas in ein CSV schreiben

```python
pd.DataFrame(final_data)
```
